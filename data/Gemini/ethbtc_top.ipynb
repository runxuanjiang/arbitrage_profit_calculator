```python
import websocket
import json
import pandas as pd
try:
    import thread
except ImportError:
    import _thread as thread
import time

coin = "ethbtc" #Set you coin
init = 1
header = {"timestamp":[], "timestampsms":[], "highestBid":[], "Bidremaining":[],"lowestAsk":[],"Askremaining":[], "midpoint":[]}

header = pd.DataFrame(header)
highestBid = 0
lowestAsk = 0

def on_message(ws, message):
    result = json.loads(message)
    lst = pd.DataFrame.from_dict(result)
    global init, coin, highestBid, lowestAsk, bremaining, aremaining

    if (init == 1):
        #file path, one for initial data, one for transaction records
        #the folder needs to be created before the program
        highestBid = lst['events'][0]['price']
        bremaining = lst['events'][0]['remaining']
        lowestAsk = lst['events'][1]['price']
        aremaining = lst['events'][1]['remaining']
        header.to_csv('1-22/' + coin + '_top.csv',header=True, index=False)
        init = 0
    else:
        if (lst['events'][0]['type']=="change"):
            side = lst['events'][0]['side']
            remaining = lst['events'][0]['remaining']
        else:
            side = lst['events'][1]['side']
            remaining = lst['events'][1]['remaining']
        if (side == 'ask'):
            lowestAsk = lst['events'][0]['price']
            aremaining = remaining
        else:
            highestBid = lst['events'][0]['price']
            bremaining = remaining

        midpoint = (float(lowestAsk) + float(highestBid))/2
        data = [[lst['timestamp'][0],lst['timestampms'][0], highestBid, bremaining, lowestAsk, aremaining,midpoint]]
        temp = pd.DataFrame(data)
        temp.to_csv('1-22/'+coin+'_top.csv', mode= 'a', header=False, index=False)
def on_error(ws, error):
    print(error)

def on_close(ws):
    print("### closed ###")

if __name__ == "__main__":
    websocket.enableTrace(True)
    #url here, add any other options according to API document
    ws = websocket.WebSocketApp("wss://api.gemini.com/v1/marketdata/"+coin+"?top_of_book=true",
                              on_message = on_message,
                              on_error = on_error,
                              on_close = on_close)
    ws.run_forever(ping_interval=5)
```
