```python
import websocket
import json
import pandas as pd
try:
    import thread
except ImportError:
    import _thread as thread
import time

coin = "btcusd" #Set you coin
init = 1
header = {"type":[],"eventId":[],"timestamp":[],"timestampms":[],"socket_sequence":[],"events":[]}
header = pd.DataFrame(header)

def on_message(ws, message):
    result = json.loads(message)
    lst = pd.DataFrame.from_dict(result)
    global init, coin
    if (init == 1):
        #file path, one for initial data, one for transaction records
        #the folder needs to be created before the program
        lst.to_csv('Gemini/orders/'+coin+'_init.csv', header=True, index=False)
        header.to_csv('Gemini/orders/'+coin+'.csv', header=True,index=False)
        init = 0
    else:
        lst.to_csv('Gemini/orders/'+coin+'.csv', mode= 'a', header=False, index=False)
def on_error(ws, error):
    print(error)

def on_close(ws):
    print("### closed ###")

if __name__ == "__main__":
    websocket.enableTrace(True)
    #url here, add any other options according to API document
    ws = websocket.WebSocketApp("wss://api.gemini.com/v1/marketdata/"+coin+"?bids=true&offers=true",
                              on_message = on_message,
                              on_error = on_error,
                              on_close = on_close)
    ws.run_forever(ping_interval=60,ping_timeout=5)
```
